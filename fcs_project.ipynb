 ## 1. Create a single dataframe with the concatenation of all nput csv files, adding a column called country

```python
import pandas as pd

files = [
    "CAvideos.csv.zst",
    "DEvideos.csv.zst",
    "FRvideos.csv.zst",
    "GBvideos.csv.zst",
    "INvideos.csv.zst",
    "JPvideos.csv.zst",
    "KRvideos.csv.zst",
    "MXvideos.csv.zst",
    "RUvideos.csv.zst",
    "USvideos.csv.zst"
]

dfs=[]

for f in files:
    country = f[:2]
    df_csv = pd.read_csv(f, compression= "zstd", encoding= "latin1")
    df_csv["country"]=country
    dfs.append(df_csv)

df_all = pd.concat(dfs,ignore_index =True)
print(df_all.shape)
df_all.head()
```

## 2. Extract all videos that have no tag.

## 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded , and remove those rows from the original dataframe.

## 5. Add a like ratio column storing the ratio between the number of likes and of dislikes

```python
df_all["likes_ratio"] = df_all["likes"]/df_all["dislikes"]
df_all.loc[df_all["dislikes"]==0, "likes_ratio"] = pd.NA
df_all.head()
```

## 6. Cluster the publish time into 10-minute intervals (e.g.from 02:20 to 02:30)


```python
# convert publish_time to datetime
df_all["publish_time"] = pd.to_datetime(df_all["publish_time"])

# floor publish_time to 10-minute intervals
df_all["publish_interval"] = df_all["publish_time"].dt.floor("10T")
df_all.head()
```

## 7. For each interval, determine the number of videos,average number of likes and of dislikes.

```python
interval_stats = (
    df_all
    .groupby("publish_interval")
    .agg(
        num_videos=("video_id", "count"),
        avg_likes=("likes", "mean"),
        avg_dislikes=("dislikes", "mean")
    )
    .reset_index()
)

interval_stats.head()
```

## 14. Read all json files with the video categories

```python
import json
files=[
    "CA_category_id.json",
    "DE_category_id.json",
    "FR_category_id.json",
    "GB_category_id.json",
    "IN_category_id.json",
    "JP_category_id.json",
    "KR_category_id.json",
    "MX_category_id.json",
    "RU_category_id.json",
    "US_category_id.json"
]
category_rows=[]

for f in files:
    country= f[:2]

    # read the json files:
    with open(f,"r",encoding="utf8") as file:
        data = json.load(file)

    # extract categories
    for item in data["items"]:
        category_rows.append({
            "category_id":int(item["id"]),
            "category_name":item["snippet"]["title"],
            "assignable":item["snippet"]["assignable"],
            "country":country
        })

categories_df = pd.DataFrame(category_rows)
categories_df.head()
    
```

## 15. For each country, determine how many videos have a category that is not assignable.

```python
df_joined = df_all.merge(
    categories_df,
    on = ["category_id","country"],
    how="inner"
)
all_countries = df_all["country"].unique()

df_not_assignable = df_joined[df_joined["assignable"] == False]

result= df_not_assignable.groupby("country").size().reindex(all_countries, fill_value=0).reset_index(name="num_videos_not_assignable")
print(result)
```
